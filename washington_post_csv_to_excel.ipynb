```python
import pandas as pd
```

```python
csv_name='washingtonpost.csv'
data=pd.read_csv(csv_name)
data.head()
```

```python
data=data[:1300]
data.shape
```

```python
def add_format(x):
    if('video' in x):
        return '视频'
    return '文字'
```

```python
data['报道形式（文字／图片／专访等）']=data['链接'].apply(lambda x:add_format(x))
# data['报道标题']=data['报道标题']+'\n'+data['链接']
```

```python
from baidu_translate import BaiduTranslate
import time
BaiduTranslate_test = BaiduTranslate('en','zh')
```

```python
def get_Chinese(x):
    Results = BaiduTranslate_test.BdTrans(x)#要翻译的词组
    time.sleep(0.9)
    print(Results)
    return Results[1]
```

```python
data['报道标题']=data['报道标题'].apply(lambda x:get_Chinese(x))
```

```python
basestation_end='The_Washington_Post.xls'
data.to_excel(basestation_end,columns=['报道时间','报道人','报道形式（文字／图片／专访等）','报道标题','链接'])
```
