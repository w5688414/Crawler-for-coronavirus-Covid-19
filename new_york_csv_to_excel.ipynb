```python
import pandas as pd
```

```python
csv_name='new_york_times.csv'
data=pd.read_csv(csv_name)
```

```python
data.head()
```

```python
data.shape
```

```python
url=data.iloc[0]['链接']
```

```python
print(url)
url=url.split('?')[0]
print(url)
```

```python
full_path='https://www.nytimes.com'+url
print(full_path)
```

```python
def fill_url(url):
    url=url.split('?')[0]
    full_path='https://www.nytimes.com'+url
    return full_path
```

```python
data['链接']=data['链接'].apply(lambda x:fill_url(x))
```

```python
data.head()
```

```python
def add_format(x):
    return '文字'
```

```python
data['报道形式（文字／图片／专访等）']=data['链接'].apply(lambda x:add_format(x))
```

```python
from baidu_translate import BaiduTranslate
BaiduTranslate_test = BaiduTranslate('en','zh')
```

```python
import time
def get_Chinese(x):
    Results = BaiduTranslate_test.BdTrans(x)#要翻译的词组
    time.sleep(0.9)
    print(Results)
    return Results[1]
```

```python
data['报道标题']=data['报道标题'].apply(lambda x:get_Chinese(x))
data.head()
```
